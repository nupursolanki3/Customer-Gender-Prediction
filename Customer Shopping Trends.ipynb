# Creating a model to predict gender based on customer information

```python
X = df.drop(['Customer ID','Gender'], axis=1)
Y = df['Gender']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state =42)
```

Standardizing Feature


```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
```

```python
X_train_tensor = torch.FloatTensor(X_train_scaled)
X_test_tensor = torch.FloatTensor(X_test_scaled)
y_train_tensor = torch.FloatTensor(y_train.values)
y_test_tensor = torch.FloatTensor(y_test.values)
```

Creating custom dataset class for pytorch

```python
class CustomDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels
    
    def __len__(self):
        return len(self.features)
    
    def __getitem__(self, idx):
        return self.features[idx], self.labels[idx]
```

Create datasets and dataloaders

```python
train_dataset = CustomDataset(X_train_tensor, y_train_tensor)
test_dataset = CustomDataset(X_test_tensor, y_test_tensor)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
```

Dictionary to store performance metrics

```python
performance_metrics = {
    'Model': [],
    'Accuracy': [],
    'F1 Score': [],
    'AUC': []
}
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train_scaled, y_train)
lr_pred = lr.predict(X_test_scaled)
lr_prob = lr.predict_proba(X_test_scaled)[:,1]
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_scaled, y_train)
rf_pred = rf.predict(X_test_scaled)
rf_prob = rf.predict_proba(X_test_scaled)[:, 1]
```

# SVM

```python
from sklearn.svm import SVC
```

```python
svm = SVC(probability=True, random_state=42)
svm.fit(X_train_scaled, y_train)
svm_pred = svm.predict(X_test_scaled)
svm_prob = svm.predict_proba(X_test_scaled)[:, 1]
```

# Evaluation

```python
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix, roc_curve, auc
)
```

```python
def evaluate_model(name, y_true, y_pred, y_proba=None):
    print(f'====={name}=====')
    print('\nClassification Reort')
    print(classification_report(y_true, y_pred))
    
    acc = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    
    performance_metrics['Model'].append(name)
    performance_metrics['Accuracy'].append(acc)
    performance_metrics['F1 Score'].append(f1)
    
    fig, axes = plt.subplots(1,2,figsize=(16,6))
    
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0])
    axes[0].set_title(f'Confusion Matrix - {name}')
    axes[0].set_xlabel('Predicted')
    axes[0].set_ylabel('Actual')
    
    roc_auc = None
    if y_proba is not None:
        fpr, tpr, _ = roc_curve(y_true, y_proba)
        roc_auc = auc(fpr, tpr)
        performance_metrics['AUC'].append(roc_auc)
        
        axes[1].plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.3f})')
        axes[1].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        axes[1].set_xlim([0.0, 1.0])
        axes[1].set_ylim([0.0, 1.05])
        axes[1].set_xlabel('False Positive Rate')
        axes[1].set_ylabel('True Positive Rate')
        axes[1].set_title(f'ROC Curve - {name}')
        axes[1].legend(loc='lower right')
    else:
        performance_metrics['AUC'].append(None)
    
    plt.tight_layout()
    plt.show()
    
    return roc_auc
```

```python
print("y_test shape:", y_test.shape)
print("dt_prob shape:", dt_prob.shape)
```

```python
lr_auc = evaluate_model("Logistic Regression", y_test, lr_pred, lr_prob)
dt_auc = evaluate_model("Decision Tree", y_test, dt_pred, dt_prob)
rf_auc = evaluate_model("Random Forest", y_test, rf_pred, rf_prob)
svm_auc = evaluate_model("Support Vector Machine", y_test, svm_pred, svm_prob)
```

# Feature Importance

```python
from sklearn.inspection import permutation_importance
import numpy as np
plt.figure(figsize=(15,10))

feature_importances = {}

if hasattr(lr, 'coef_'):
    lr_importance = np.abs(lr.coef_[0])
    feature_importances['Logistic Regression'] = lr_importance
    
if hasattr(dt, 'feature_importances_'):
    dt_importance = dt.feature_importances_
    feature_importances['Decision Tree'] = dt_importance
    
if hasattr(rf, 'feature_importances_'):
    rf_importance = rf.feature_importances_
    feature_importances['Random Forest'] = rf_importance
    
result = permutation_importance(svm, X_test_scaled, y_test, n_repeats=10, random_state=42)
svm_importance = result.importances_mean
feature_importances['SVM'] = svm_importance

for i, (model_name, importance) in enumerate(feature_importances.items()):
    plt.figure(figsize=(10, 6))
    importance_df = pd.DataFrame({
        'Feature': X.columns,
        'Importance': importance
    }).sort_values('Importance', ascending=False)
    
    sns.barplot(x='Importance', y='Feature', data=importance_df)
    plt.title(f'Feature Importance - {model_name}')
    plt.tight_layout()
    plt.show()
    
    print(f"\n{model_name} Feature Importance:")
    print(importance_df)
    
importance_comparison = pd.DataFrame(
    {model_name: imp for model_name, imp in feature_importances.items()},
    index=X.columns
)

for column in importance_comparison.columns:
    max_val = importance_comparison[column].max()
    if max_val > 0:  # Avoid division by zero
        importance_comparison[column] = importance_comparison[column] / max_val
        
importance_comparison['Average'] = importance_comparison.mean(axis=1)
importance_comparison = importance_comparison.sort_values('Average', ascending=False)
importance_comparison = importance_comparison.drop('Average', axis=1)

plt.figure(figsize=(12, 10))
sns.heatmap(importance_comparison, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Feature Importance Comparison Across Models')
plt.tight_layout()
plt.show()
```

```python
from IPython.display import display
# Create the performance summary table with highlighted best scores
performance_df = pd.DataFrame(performance_metrics)

# Function to highlight the best value in each metric column
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: blue' if v else '' for v in is_max]

# Apply styling to highlight best scores
styled_performance = performance_df.style.apply(highlight_max, subset=['Accuracy', 'F1 Score', 'AUC'])

# Display the styled summary table
print("\n===== Model Performance Summary =====")
display(styled_performance)

# Also create a visual representation of the performance metrics
plt.figure(figsize=(14, 8))

# Create a grouped bar chart
metrics = ['Accuracy', 'F1 Score', 'AUC']
x = np.arange(len(performance_df['Model']))
width = 0.25

fig, ax = plt.subplots(figsize=(14, 8))
rects1 = ax.bar(x - width, performance_df['Accuracy'], width, label='Accuracy')
rects2 = ax.bar(x, performance_df['F1 Score'], width, label='F1 Score')
rects3 = ax.bar(x + width, performance_df['AUC'], width, label='AUC')

# Add labels and formatting
ax.set_ylabel('Score')
ax.set_title('Performance Metrics by Model')
ax.set_xticks(x)
ax.set_xticklabels(performance_df['Model'], rotation=45, ha='right')
ax.legend()
ax.set_ylim(0, 1.0)

# Add value labels on top of bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        if not np.isnan(height):  # Skip if value is NaN
            ax.annotate(f'{height:.3f}',
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom', fontsize=8)

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.tight_layout()
plt.show()
```
